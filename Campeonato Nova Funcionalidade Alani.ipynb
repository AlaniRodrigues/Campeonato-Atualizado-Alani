```python
print('Vamos começar o Campeonato de Tênis!')
print('Primeiro vamos começar pelo Cadastro dos times e Jogadores!')

def cadastrar_times():
    times = []
    while len(times) < 4:
        nome_time = input(f"Digite o nome do Time\n(ou 'fim' para encerrar): ")
        if nome_time.lower() == "fim":
            break

        jogadores = []
        nomes_jogadores = []

        while len(jogadores) < 4:
            nome_jogador = input(f"Digite o nome do jogador {len(jogadores) + 1}: ")
            if nome_jogador in nomes_jogadores:
                print("Jogador já cadastrado neste time.\Tente outro nome.")
                continue

            posicao = input(f"Digite a posição do jogador {len(jogadores) + 1}: ")
            jogador = (nome_jogador, posicao)
            jogadores.append(jogador)
            nomes_jogadores.append(nome_jogador)

        time = {
            'nome': nome_time,
            'jogadores': jogadores,
            'pontuacao': 0
        }

        times.append(time)

    return times

def realizar_confronto(time1, time2):
    print(f"\nConfronto entre {time1['nome']} vs {time2['nome']}")
    p1 = int(input(f"Digite a pontuação de {time1['nome']}: "))
    p2 = int(input(f"Digite a pontuação de {time2['nome']}: "))
    time1['pontuacao'] += p1
    time2['pontuacao'] += p2
    if p1 > p2:
        return time1
    elif p2 > p1:
        return time2
    else:
        print("Empate! O confronto será repetido.")
        return realizar_confronto(time1, time2)

def campeonato(times):
    rodada = 1
    while len(times) > 1:
        print(f"\n Iniciando a Rodada {rodada}")
        vencedores = []
        i = 0
        while i < len(times) - 1:
            time1 = times[i]
            time2 = times[i + 1]
            vencedor = realizar_confronto(time1, time2)
            vencedores.append(vencedor)
            i += 2
        if len(times) % 2 != 0:
            print(f"{times[-1]['nome']} avançou automaticamente!")
            vencedores.append(times[-1])
        times = vencedores
        rodada += 1

    campeao = times[0]
    print(f"\nO campeão do torneio é {campeao['nome']} com {campeao['pontuacao']} pontos!")

times_cadastrados = cadastrar_times()
if len(times_cadastrados) < 2:
    print("É necessário pelo menos 2 times para iniciar o campeonato.")
else:
    campeonato(times_cadastrados)

from collections import deque

fila_reclamacoes = deque()
registro_reclamacoes = {}

def registrar_reclamacao(times_validos):
    nome_time = input("Digite o nome do time que deseja reclamar: ")

    nomes_times = [time['nome'] for time in times_validos]
    if nome_time not in nomes_times:
        print("⚠️ Este time não participou do campeonato. Tente novamente.")
        return

    motivo = input("Descreva o motivo da reclamação: ")

    reclamacao = {'time': nome_time, 'motivo': motivo}
    fila_reclamacoes.append(reclamacao)

    if nome_time in registro_reclamacoes:
        registro_reclamacoes[nome_time] += 1
    else:
        registro_reclamacoes[nome_time] = 1

    print("✅ Reclamação registrada com sucesso!")

def processar_reclamacao():
    if fila_reclamacoes:
        reclamacao = fila_reclamacoes.popleft()
        print(f"📣 Processando reclamação do time {reclamacao['time']}: {reclamacao['motivo']}")
    else:
        print("Nenhuma reclamação para processar.")

def ver_total_por_time():
    if not registro_reclamacoes:
        print("Nenhum time fez reclamações.")
        return
    print("\n--- TOTAL DE RECLAMAÇÕES POR TIME ---")
    for time, total in registro_reclamacoes.items():
        print(f"{time}: {total} reclamação")

def menu(times_validos):
    while True:
        print("\n📋 MENU DE RECLAMAÇÕES")
        print("1. Registrar nova reclamação")
        print("2. Processar próxima reclamação")
        print("3. Ver total por time")
        print("4. Sair")
        opcao = input("Escolha uma opção: ")

        if opcao == "1":
            registrar_reclamacao(times_validos)
        elif opcao == "2":
            processar_reclamacao()
        elif opcao == "3":
            ver_total_por_time()
        elif opcao == "4":
            print("Saindo do sistema de reclamações.")
            break
        else:
            print("Opção inválida. Tente novamente.")

menu(times_cadastrados)
```
